```python
# import libraries

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# set theme
sns.set_theme(style="darkgrid")
```

## Loading Data Sets in Seaborn

```python
# Seaborn has built in data sets

# Loading a sample dataset (tips)
tips = sns.load_dataset('tips')

# Viewing the data
tips.head()
```

```python
# Loading a sample dataset (iris)
iris = sns.load_dataset('iris')

# Viewing the data
iris.head()
```

```python
# Loading a sample dataset (titanic)
titanic = sns.load_dataset('titanic')

# Viewing the data
titanic.head()
```

```python
# load dataset from sklearn
from sklearn.datasets import load_boston

housing = load_boston()

house = pd.DataFrame(housing.data, columns=housing.feature_names)
house['PRICE'] = housing.target
house.head()
```

## Scatter Plot
#### Scatter plots are used for inspection of relationship between two numeric variables

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue= 'smoker', style = 'smoker', col = 'time', size = 'size', kind = 'scatter')
```

```python
sns.scatterplot(x = 'sepal_length', y = 'petal_length', hue = 'species', data = iris)
```

```python
sns.scatterplot(x = 'sepal_width', y = 'petal_width', hue = 'species', data = iris)
```

```python
sns.countplot(data = titanic, x = 'class')
```

```python
sns.countplot(data = titanic, x = 'survived')
```

## Bar Chart

### Used for representing the distribution of categorical data

```python
sns.barplot(data = titanic, x = 'sex', y = 'survived', hue = 'class')
```

```python
sns.displot(house['PRICE'])
```

## Heat Map

### A heat map is 2D version of a histrogram, an alternative approach to scatter plots.

```python
correlation = house.corr()
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(correlation, cbar = True, square = True, fmt = '.1f', annot = True, annot_kws = {'size':8}, cmap = 'Blues')
```
