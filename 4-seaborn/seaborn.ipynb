```python
# import libraries

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# set theme
sns.set_theme(style="darkgrid")
```

## Loading Data Sets in Seaborn

```python
# Seaborn has built in data sets

# Loading a sample dataset (tips)
tips = sns.load_dataset('tips')

# Viewing the data
tips.head()
```

```python
# Loading a sample dataset (iris)
iris = sns.load_dataset('iris')

# Viewing the data
iris.head()
```

```python
# Loading a sample dataset (titanic)
titanic = sns.load_dataset('titanic')

# Viewing the data
titanic.head()
```

```python
# load dataset from sklearn
from sklearn.datasets import load_boston

housing = load_boston()

house = pd.DataFrame(housing.data, columns=housing.feature_names)
house['PRICE'] = housing.target
house.head()
```

# Univariate Visualization

## Visualizations for single variable

### Count Plot

##### A count plot can be thought of as a histogram across a categorical, instead of quantitative, variable.

```python
sns.countplot(data = titanic, x = 'class')
```

```python
sns.countplot(data = titanic, x = 'survived')
```

### Histrogram

##### A histrogram is used to plot the distribution of a numeric variable

```python
sns.displot(house['PRICE'])
```

# Bivariate Visualization

## Visualizations for inspecting relationship between two variables.

### Scatter Plot

##### Scatter plots are used for inspection of relationship between two numeric variables

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', col = 'time', kind = 'scatter')
```

```python
sns.scatterplot(x = 'sepal_length', y = 'petal_length', data = iris)
```

```python
sns.scatterplot(x = 'sepal_width', y = 'petal_width', data = iris)
```

### Adaptive Bar Plot

##### A bar plot represents an estimate of central tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainty around that estimate using error bar.

```python
sns.barplot(data = titanic, x = 'sex', y = 'survived', hue = 'class')
```

### Point Plot

##### Point plots can be more useful than bar plots for focusing comparisons between different levels of one or more categorical variables. They are particularly adept at showing interactions: how the relationship between levels of one categorical variable changes across levels of a second categorical variable. The lines that join each point from the same hue level allow interactions to be judged by differences in slope, which is easier for the eyes than comparing the heights of several groups of points or bars.

```python
sns.pointplot(data = titanic, x = 'sex', y = 'survived', hue = 'class')
```

### Heat Map

##### A heat map is 2D version of a histrogram, an alternative approach to scatter plots.

```python
correlation = house.corr()
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(correlation, cbar = True, square = True, fmt = '.1f', annot = True, annot_kws = {'size':8}, cmap = 'Blues')
```

### Violin Plot

##### Used for plotting relationship between quantative and qualitative variables.

```python
sns.violinplot(data = tips, x = 'sex', y = 'total_bill')
```

### Box Plot

##### Another way of plotting relationship between quantative and qualitative variables.

```python
sns.boxplot(data = tips, x = 'sex', y = 'total_bill')
```

### Clustered Bar Chart

##### Used for depicting a relationship between two categorical variables.

```python
sns.countplot(data = tips, x = 'sex', hue = 'time')
```

# Multivariate Visualization

## Visualizations for inspecting relationship between more than two variables.

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', hue= 'smoker', style = 'smoker', col = 'time', size = 'size', kind = 'scatter')
```

```python
sns.scatterplot(x = 'sepal_length', y = 'petal_length', hue = 'species', data = iris)
```

```python
sns.scatterplot(x = 'sepal_width', y = 'petal_width', hue = 'species', data = iris)
```
