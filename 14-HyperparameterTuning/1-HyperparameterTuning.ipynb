## Import Dependencies

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

## Data and Preprocessing

```python
# load data from sklearn
breast_cancer_data = sklearn.datasets.load_breast_cancer()
print(breast_cancer_data)
```

```python
my_data = pd.DataFrame(breast_cancer_data.data, columns=breast_cancer_data.feature_names)
my_data.head()
```

```python
# Adding target column

my_data['target'] = breast_cancer_data.target
my_data.head()
```

```python
# shape of data

my_data.shape
```

```python
# checking for null values

my_data.isnull().sum()
```

```python
my_data.target.value_counts()
```

```python
# Separating the feature and target

X = my_data.drop(columns='target', axis=1)
Y = my_data.target
```

```python
X = np.asarray(X)
Y = np.asarray(Y)
```

## GridSearchCV

```python
# load the model

model = SVC()
```

```python
# Hyperparameters

parameters = {
    'kernel':['linear', 'poly', 'rbf', 'sigmoid'],
    'C': [1,5,10,15,20]
}
```

```python
# gridsearch

classifier = GridSearchCV(model, parameters, cv = 5)
```

```python
# fit data to the model

classifier.fit(X,Y)
```

```python
classifier.cv_results_
```

```python
# best parameters

classifier.best_params_
```

```python
# highest accuracy

classifier.best_score_
```

```python
# load results to a dataframe

result = pd.DataFrame(classifier.cv_results_)
result.head()
```

```python
# gridsearch result

gridsearch_result = result[['param_C', 'param_kernel', 'mean_test_score']]
gridsearch_result.head()
```
