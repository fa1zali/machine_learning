Label Encoding: Converting lables into numeric form

```python
# import libraries

import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding for Breast Cancer Dataset

```python
breast_cancer_df = pd.read_csv('datasets/breast_cancer_data.csv')

breast_cancer_df.head()
```

```python
# find count of different lablels

breast_cancer_df['diagnosis'].value_counts()
```

```python
# load label encoder

encoder = LabelEncoder()
```

```python
labels = encoder.fit_transform(breast_cancer_df.diagnosis)
```

```python
# adding labels column
breast_cancer_df['target'] = labels
```

```python
breast_cancer_df.head()
```

```python
breast_cancer_df['target'].value_counts()
```

Label Encoding for IRIS Dataset

```python
iris_df = pd.read_csv('datasets/iris_data.csv')

iris_df.head()
```

```python
iris_df['Species'].value_counts()
```

```python
iris_encoder = LabelEncoder()
```

```python
iris_label = iris_encoder.fit_transform(iris_df.Species)
```

```python
iris_df['target'] = iris_label
```

```python
iris_df.head()
```

```python
iris_df['target'].value_counts()
```
